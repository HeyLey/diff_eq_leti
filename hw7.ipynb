```python
import numpy as np

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python

#   x'' − µ(1 − x^2) x' + x = f(t)

def solve(x_0, dx_0, mu, A, omega, max_steps = 100000):
    t = np.zeros(max_steps)
    x_all = np.zeros(max_steps)
    dx_all = np.zeros(max_steps)
    ddx_all = np.zeros(max_steps)

    
    current_t = 0.0
    
    x = x_0
    dx = dx_0
        
    for i in range(max_steps):
        ddx = mu * (1 - x ** 2) * dx - x + A * np.sin(omega * current_t) 
        
        
        t[i] = current_t
        x_all[i] = x
        dx_all[i] = dx
        ddx_all[i] = ddx
        
        step = 0.001 
        x += dx * step
        dx += ddx * step
        
        current_t += step
        
    return t[:i], x_all[:i], dx_all[:i], ddx_all[:i]

    
```

Вынужденные колебания 

```python
t, x_all, dx_all, ddx_all = solve(
    x_0 = 0.0, 
    dx_0 = 0.0, 
    mu = 1, 
    A = 1, 
    omega = 1,
    max_steps = 20000)

plt.plot(t, x_all, label="x")
plt.plot(t, dx_all, label="x'")
plt.plot(t, ddx_all, label="x''")
plt.legend()
```
