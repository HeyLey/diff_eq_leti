```python
import numpy as np

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solve(A_0, B_0, C_0, k1, k2, k3, max_t = 1000, max_steps = 100000):
    A, B, C = A_0, B_0, C_0
    
    t = np.zeros(max_steps)
    a_all = np.zeros(max_steps)
    b_all = np.zeros(max_steps)
    c_all = np.zeros(max_steps)
    da_all = np.zeros(max_steps)
    db_all = np.zeros(max_steps)
    dc_all = np.zeros(max_steps)
            
    current_t = 0.0
        
    for i in range(max_steps):
        da = -k1 * A + k2 * B
        db = k1 * A - (k2 + k3) * B
        dc = k3 * B
        
        t[i] = current_t
        a_all[i] = A
        b_all[i] = B
        c_all[i] = C
        da_all[i] = da
        db_all[i] = db
        dc_all[i] = dc
        
        
        m = max(abs(da), abs(db), abs(dc))
        step = 0.01 / max(m, 0.1)
        A += da * step
        B += db * step
        C += dc * step
        current_t += step
        if m < 0.001:
            break
        if current_t >= max_t:
            break
    return t[:i], a_all[:i], b_all[:i], c_all[:i], da_all[:i], db_all[:i], dc_all[:i]

    
```

```python
t, a_all, b_all, c_all, da_all, db_all, dc_all = solve(
    A_0 = 1.0, 
    B_0 = 1.0, 
    C_0 = 1.0, 
    k1 = 1.0, 
    k2 = 1.0, 
    k3 = 1.0, 
    max_t=20)

f, axarr = plt.subplots(2)

axarr[0].plot(t, a_all, label="A")
axarr[0].plot(t, b_all, label="B")
axarr[0].plot(t, c_all, label="C")
axarr[0].legend()

axarr[1].plot(t, da_all, label="dA")
axarr[1].plot(t, db_all, label="dB")
axarr[1].plot(t, dc_all, label="dC")
axarr[1].legend()
```
