```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

%matplotlib inline 
```

```python
#1

Y, X = np.mgrid[-15:15:0.1, -15:15:0.1]
dx = X - 3*Y - 1
dy = 6*X - 10*Y - 14

plt.figure(figsize=(10,10))
plt.streamplot(X, Y, dx, dy)

plt.plot([4.0], [1.0], "ro")
plt.arrow(4, 1, 4 + 3, 1 + 2, head_width=0.1, head_length=0.3, fc='k', ec='k')
plt.arrow(4, 1, 4 + 1, 1 + 3, head_width=0.1, head_length=0.3, fc='k', ec='k')
```

```python
#2

Y, X = np.mgrid[-15:15:0.1, -15:15:0.1]
dx = 3*X + 2*Y - 1
dy = (-9)*X - 3*Y + 15

plt.figure(figsize=(10,10))
plt.streamplot(X, Y, dx, dy)

plt.plot([3.0], [-4.0], "ro")
```

```python
#3

Y, X = np.mgrid[-15:15:0.1, -15:15:0.1]
dx = 6*X + Y + 8
dy = (-8)*X - 3*Y - 4

plt.figure(figsize=(10,10))
plt.streamplot(X, Y, dx, dy)

plt.plot([-2.0], [4.0], "ro")
plt.arrow(-2, 4, -2 - 1, 4 + 1, head_width=0.1, head_length=0.3, fc='k', ec='k')
plt.arrow(-2, 4, -2 - 1, 4 + 8, head_width=0.1, head_length=0.3, fc='k', ec='k')
```
